## Q.3

App routing in flask refers to the process of mapping URLs to corresponding functions in a Flask application.

App routes can be used like a decorator to define routes. The decorator is applied to a function and the function is executed when a request is made to the ssociated URL.

syntax: @app.route('/')

```python
# Q.4
# Create a “/welcome” route to display the welcome message “Welcome to ABC Corporation” and a “/”
#route to show the following details:
#Company Name: ABC Corporation
#Location: India
#Contact Detail: 999-999-9999

from flask import Flask

app = Flask(__name__)

@app.route("/")
def details():
    Company_Name = 'ABC Corporation'
    Location = 'India'
    Contact_Detail = '999-999-9999'
    return f"Company Name: {Company_Name} <br> Location: {Location} <br> Contact Detail: {Contact_Detail}"

@app.route('/welcome')
def welcome():
    return "<H2> Welcome to ABC Corporation </H2>"


if __name__=="__main__":
    app.run(host="0.0.0.0")
```

```python
# Q.5

# In Flask, the url_for() function is used for URL building.

from flask import Flask, url_for

app = Flask(__name__)

@app.route('/')
def index():
    return 'Home Page'

@app.route('/login')
def login():
    return 'Login Page'

@app.route('/logout')
def logout():
    return 'Logout Page'

@app.route('/dashboard')
def dashboard():
    home_url = url_for('index')
    login_url = url_for('login')
    logout_url = url_for('logout')

    return f'''
        <p>Home: <a href="{home_url}">{home_url}</a></p>
        <p>Login: <a href="{login_url}">{login_url}</a></p>
        <p>Logout: <a href="{logout_url}">{logout_url}</a></p>
    '''

if __name__ == '__main__':
    app.run()
```
